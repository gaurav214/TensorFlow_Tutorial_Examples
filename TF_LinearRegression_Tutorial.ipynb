## A linear regression learning algorithm example using TensorFlow library.
## Author: Gaurav Negi

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
learning_rate = 0.01
training_epochs = 2000
display_step = 50
```

### Generating Data

```python
train_X = np.asarray([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,
                         7.042,10.791,5.313,7.997,5.654,9.27,3.1])
train_Y = np.asarray([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,
                         2.827,3.465,1.65,2.904,2.42,2.94,1.3])
n_samples = train_X.shape[0]
```

### Set Tensorflow Graph

```python
X = tf.placeholder(tf.float32, name="input")
Y = tf.placeholder(tf.float32, name="output")

#randomly setting weight and bias
W = tf.Variable(np.random.randn(), name="weight")
b = tf.Variable(np.random.randn(), name="bias")
```

### Construct a Linear model y = mx +b where W denotes weight and b is bias

```python
pred = tf.add(tf.multiply(X, W), b)
```

### Caculating cost function

```python
cost = tf.reduce_sum(tf.pow(pred-Y, 2))/(2*n_samples)
```

### Gradient descent, minimize() knows to modify W and b because Variable objects are trainable=True by default

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

### Intialize all the varaibles with default values

```python
init = tf.global_variables_initializer()
```

### Start Training

```python
sess = tf.Session()
#initialize

sess.run(init)

#Fit all the training data
for epoch in range(training_epochs):
    for x,y in zip(train_X, train_Y):
        sess.run(optimizer, feed_dict = {X: x, Y:y}) #feeding values to placeholders
        
        #Display each epoch step
        if(epoch + 1) % display_step == 0:
            c = sess.run(cost, feed_dict = {X: train_X, Y: train_Y})
            print("Epoch: ", '%4d' % (epoch + 1), " cost =", "{:.9f}".format(c), "W =", sess.run(W), "B =",sess.run(b),'\n')
    
```

```python
 print("Optimization Finished!")
 training_cost = sess.run(cost, feed_dict={X: train_X, Y: train_Y})
 print("Training cost=", training_cost, "W=", sess.run(W), "b=", sess.run(b), '\n')
```

### Visualize

```python
 # Graphic display
plt.plot(train_X, train_Y, 'ro', label='Original data')
plt.plot(train_X, sess.run(W) * train_X + sess.run(b), label='Fitted line')
plt.legend()
plt.show()
```

### Testing

```python
test_X = numpy.asarray([6.83, 4.668, 8.9, 7.91, 5.7, 8.7, 3.1, 2.1])
test_Y = numpy.asarray([1.84, 2.273, 3.2, 2.831, 2.92, 3.24, 1.35, 1.03])
```

```python
print("Testing... (Mean square loss Comparison)")
testing_cost = sess.run(
        tf.reduce_sum(tf.pow(pred - Y, 2)) / (2 * test_X.shape[0]),
        feed_dict={X: test_X, Y: test_Y})  # same function as cost above
print("Testing cost=", testing_cost)
print("Absolute mean square loss difference:", abs(
        training_cost - testing_cost))
```

```python
plt.plot(test_X, test_Y, 'bo', label='Testing data')
plt.plot(train_X, sess.run(W) * train_X + sess.run(b), label='Fitted line')
plt.legend()
plt.show()
```
